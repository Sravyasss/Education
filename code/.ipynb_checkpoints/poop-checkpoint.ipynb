```python
cols_to_drop = [
'SCH_FTETEACH_CERT',
'SCH_FTETEACH_NOTCERT',
'SCH_FTETEACH_FY',
'SCH_FTETEACH_SY',
'SCH_TEACHERS_CURR_TOT',
'SCH_TEACHERS_PREV_TOT',
'SCH_FTETEACH_ABSENT',
'SCH_FTECOUNSELORS',
'SCH_FTESECURITY_LEO',
'SCH_FTESECURITY_GUA',
'SCH_FTESERVICES_NUR',
'SCH_FTESERVICES_PSY',
'SCH_FTESERVICES_SOC'

]

df_edu = df.drop(columns=cols_to_drop)
df_edu.head(3)
```

```python
df_edu.describe()
```

```python
df_edu.loc[df['percent_lunch'] < 0, 'percent_lunch'] = np.nan
df_edu.loc[df['average_act'] < 1, 'average_act'] = np.nan
```

```python
predictor_variables = [
    'rate_unemployment',
    'percent_college',
    'percent_married',
    'median_income',
    'average_act',
    'percent_lunch',
    'state',
    'charter',
    'Tot_FTE_teachers'
]
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
numerical_predictors = df_edu[predictor_variables].select_dtypes(include="number").columns.to_list()
print(numerical_predictors)
```

```python
imputer.fit(df_edu.loc[:,numerical_predictors])
df_edu.loc[:, numerical_predictors] = imputer.transform(df_edu.loc[:, numerical_predictors])
```

```python
categorical_predictors = ['LEA_STATE', 'LEA_STATE_NAME', 'LEA_NAME', 'school_id', 'SCH_NAME', 'zip_code', 'id_support', 'JJ']
for col in categorical_predictors:
    df_edu[col] = df_edu[col].fillna(df_edu[col].mode()[0])

df_edu.isna().sum().to_frame(name='Number of Missing Values')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
```

```python
df_edu=pd.read_csv('C:/Users/SRAVYASRI/Downloads/education_clean.csv')
```

```python
df_edu.head()
```

```python
predictor_variables = [
    "rate_unemployment",
    "percent_college",
    "percent_married",
    "median_income",
    "percent_lunch",
    "Tot_FTE_teachers"
    
]

numerical_predictors = df_edu[predictor_variables].select_dtypes(include='number').columns.to_list()

corr_matrix = df_edu[numerical_predictors + ['average_act']].corr()

plt.figure(figsize=(10, 5))
sns.heatmap(corr_matrix, vmax=1, vmin=-1, square=True, annot=True, cmap="viridis")
plt.tick_params(labelsize=12)
plt.show()
```

```python
fig = sns.pairplot(
    data=df_edu,
    vars=numerical_predictors + ['average_act'],
    hue='charter',
    kind="reg",
    plot_kws={
        "scatter_kws": {"alpha": 0.5, "color": "k", "s": 7}
    }
)

for ax in fig.axes.flatten():
    if ax.get_xlabel() == 'CT Median Household Income':
        ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    ax.set_xlabel(ax.get_xlabel(), fontsize=8, rotation=30, ha='right')
    ax.set_ylabel(ax.get_ylabel(), fontsize=8)
    plt.setp(ax.get_xticklabels(), rotation=30, ha='right')

plt.show()
```

```python
plt.figure(figsize=(10, 3))
fractions = list(numerical_predictors)
fractions.remove('median_income')  
sns.boxplot(data=df_edu[fractions], color='k')
plt.ylabel('Proportion', fontsize=15)
plt.tick_params(labelsize=12)

plt.show()
```
